### What is the distribution  of preferred foot?


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C://Users//Admin//Desktop//New folder//MIUI//Footballer.csv")
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
foot_distribution = df['Preferred Foot'].value_counts()
foot_distribution
```

```python
plt.figure(figsize=(6, 4))
foot_distribution.plot(kind='bar')
plt.title("Preferred Foot Distribution")
plt.xlabel("Preferred Foot")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

### How can we visualize the age distribution of players?

```python
age=df['Age'].value_counts()
plt.figure(figsize=(8, 6))
plt.hist(age, bins=20, color='blue', edgecolor='black')
plt.title("Age Distribution of Players")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

### What is the international reputation of left and right foot?

```python
mean_reputation_by_foot = df.groupby('LF')['RF'].value_counts()
mean_reputation_by_foot.plot(kind='bar', figsize=(10, 6))
plt.title("Nationality Distribution by Preferred Foot")
plt.xlabel("Preferred Foot")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Nationality", bbox_to_anchor=(1, 1))
plt.show()

```

### What is the different work rate of the Players?

```python
work_rate_counts = df['Work Rate'].value_counts()
plt.figure(figsize=(10, 6))
work_rate_counts.plot(kind='bar')
plt.title("Distribution of Work Rates")
plt.xlabel("Work Rate")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.show()
```

### Top 10 countries with most number of players

```python
country_counts = df['Nationality'].value_counts()

top_countries = country_counts.head(10)

plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar')
plt.title("Top 10 Countries with Most Players")
plt.xlabel("Country")
plt.ylabel("Number of Players")
plt.xticks(rotation=45, ha="right")
plt.show()
```

### What is the work rate of players segregated by preferred foot?

```python
work_rate_by_foot = df.groupby(['Preferred Foot', 'Work Rate'])

plt.figure(figsize=(12, 6))
work_rate_by_foot.plot(kind='bar', stacked=True)
plt.title("Work Rate of Players Segregated by Preferred Foot")
plt.xlabel("Preferred Foot")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Work Rate")
plt.show()
```

### Number of players at different positions? 



```python
position_counts = df['Position'].value_counts()


plt.figure(figsize=(12, 6))
position_counts.plot(kind='bar')
plt.title("Number of Players at Different Positions")
plt.xlabel("Position")
plt.ylabel("Number of Players")
plt.xticks(rotation=45, ha="right")
plt.show()
```

### Age distribution in top countries

```python
top_countries = df['Nationality'].value_counts().head(10).index.tolist()

top_countries_data = df[df['Nationality'].isin(top_countries)]


plt.figure(figsize=(12, 6))
for country in top_countries:
    country_data = top_countries_data[top_countries_data['Nationality'] == country]
    plt.hist(country_data['Age'], bins=20, alpha=0.6, label=country)

plt.title("Age Distribution of Players in Top Countries")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()
plt.show()
```

### What is the distribution of overall score of players from top countries?

```python
top_countries = df['Nationality'].value_counts().head(10).index.tolist()

top_countries_data = df[df['Nationality'].isin(top_countries)]

plt.figure(figsize=(12, 6))
for country in top_countries:
    country_data = top_countries_data[top_countries_data['Nationality'] == country]
    plt.hist(country_data['Overall'], bins=20, alpha=0.6, label=country)

plt.title("Overall Score Distribution of Players in Top Countries")
plt.xlabel("Overall Score")
plt.ylabel("Count")
plt.legend()
plt.show()
```
